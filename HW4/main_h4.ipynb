```python
'''
PLEASE UPLOAD "util.py" FILE ON COLAB MANUALLY TO RUN CONFUSION MATRIX
'''

import numpy as np
import scipy
import matplotlib.pyplot as plt
from keras.datasets import mnist
# from util import func_confusion_matrix


# load (downloaded if needed) the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()


# transform each image from 28 by28 to a 784 pixel vector
pixel_count = x_train.shape[1] * x_train.shape[2]
x_train = x_train.reshape(x_train.shape[0], pixel_count).astype('float32')
x_test = x_test.reshape(x_test.shape[0], pixel_count).astype('float32')

# normalize inputs from gray scale of 0-255 to values between 0-1
x_train = x_train / 255
x_test = x_test / 255

# Please write your own codes in responses to the homework assignment 4

split_size = 50000

# Training set
sub1 = x_train[:split_size]
sub1_labels = y_train[:split_size]

# Validation set
sub2 = x_train[split_size:]
sub2_labels = y_train[split_size:]

```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 


#Paramater set #1
rate = 0.001 #learning rate
neu = 256 #Number of neurons
act_fnc = 'logistic' #sigmoid activation
out_fnc = 'softmax' #softmax output activation

#Define model   2 hidden layers
model = MLPClassifier(hidden_layer_sizes = (neu, neu), 
                      learning_rate_init = rate, 
                      activation = act_fnc)

#Train model using subset of size 50,000
model.fit(sub1, sub1_labels)


#define output function
model.out_activation_ = out_fnc

#Predict over validation set 
pred = model.predict(sub2)

# execute the util.py file

execfile('./util.py')

matrix, accuracy, recall, preci = func_confusion_matrix(sub2_labels, pred)

#Remove deprecation warnings


#Display confusion matrix
print(matrix)
print("\n")

#Display Accuracy, per-class precision and recall
print("Accuracy: {}\n".format(accuracy))
print("Per-class recall rate: {}\n".format(recall))
print("Per-class precision rate: {}\n".format(preci))

```
