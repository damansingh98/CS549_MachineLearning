```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
ax = plt.subplot(111)

x = np.arange(0.0, 5.0, 0.01)
y = 30+(0.5*x) 

plt.plot(x, y, lw=2)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Straight Line")
plt.grid()
```

```python
a2 = plt.subplot(111)

x2 = np.arange(10, 40, 0.01)
y2 = ((x2-25)**2) + 20

plt.plot(x2, y2, lw=2, color = 'orange')

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Quadratic Function")
plt.grid()
```

```python
warnings.filterwarnings('ignore')

a3 = plt.subplot(111)

x3 = np.arange(-20, 20, 0.01)
y3 = -np.log(x3)
y4 = -np.log(1-x3)

plt.plot(x3, y3, lw=2)
plt.plot(x3, y4, lw=2)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Log Function")
plt.grid()
```

```python
a4 = plt.subplot(111)

x4 = np.arange(0, 19, 0.01)
y5 = 1/(1+np.exp(-x4))

plt.plot(x4, y5, lw=2, color = 'red')

plt.ylim(0.4, 2)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Log Function")
plt.grid()
```

**Scenario:** I have a friend who has reached 100k subscribers on his YouTube channel. He has planned a giveaway for a couple of his subscribers. He is confused about whether he should pick someone randomly or pick someone of his choice. He wants to make sure that the person who gets his giveaway loves his youtube content. As a machine learning student, I want to help my friend classify a subscriber as a "favorite" or "unfavorite." for the giveaway. I will shortlist subscribers who have passed some constraints, and the YouTuber can pick his subscriber either randomly or by his choice.

**Goal:** classify a subscriber as favorite or unfavorite.

**Inputs:** a dataset of youtube accounts. With a few accounts already labeled as favorite or unfavorite. The size of the dataset for this task varies by the amount of comments. We consider a few features for this problem: frequency of each comment on a video, account name, number of comments on other videos, and frequency of an account name to appear on the video for the last 90 days.

**Output:** "favorite" or "unfavorite"


**Data collection:** we will use goolge's API for youtube data to extract all comments and associated accounts from each video.

**Training data:** We will use 70% of the extracted data for training purposes. We take in account the features mentioned above.  

**Validation data:** we will take about 15% of the extracted data and use it for validation purposes and check if the trained model is actually producing the right output. If a subscriber has higher values for the features it more likely to fall into the category of favorite. Also, we can make sure to check for any missing data and fullfill with N/A value.   

**Test data:** The rest of the 15% of the data will be used to test the trained model and manually check whether the output is aligns with the expected output. Atlast we can create a table of subscribers who fall into the category of "favorite" and let the youtuber choose the person for the giveaway.

**Ground-truth label:** I want the classification model to output a value of "favorite" or "not favorite," and depending on the number of accounts securing a high value in features; we will manually take some videos and look for genuine subscribers and list them as favorite or not favorite, and this will generate ground truth labels for all the dataset. Lastly, we feed the dataset to the model and expect accurate output predictions against our manual labeling.
